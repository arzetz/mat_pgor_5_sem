**Задание №1, Вариант 4.**
Разделение на 5 категорий,
данные:
[5, 10, 15, 30, 40]

```python
import matplotlib.pyplot as plt

data = [
    ['Категория 1', 5],
    ['Категория 2', 10],
    ['Категория 3', 15],
    ['Категория 4', 30],
    ['Категория 5', 40]
]

categories = [row[0] for row in data]
sales = [row[1] for row in data]

plt.figure(figsize=(8, 8))
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Разделение данных по категориям')
plt.show()
```

**Задание №2. Вариант 4.** Спираль Ферма

```python
import numpy as np

import matplotlib.pyplot as plt




def spiral_ferma(num_points, a=1):
    koef = np.linspace(0, 4 * np.pi, num_points)  #Угол от 0 до 4π
    r = a * np.sqrt(koef)
    x = r * np.cos(koef)
    y = r * np.sin(koef)
    return x, y

num_points = 500 #Количество точек для визуализации
x, y = spiral_ferma(num_points)

plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.title("Спираль Ферма")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.axis('equal')
plt.grid()
plt.show()
```
